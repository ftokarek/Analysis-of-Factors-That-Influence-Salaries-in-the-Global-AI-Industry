# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from scipy import stats
```

# Ensure results directory exists

```python
os.makedirs('../python_results', exist_ok=True)
```

# Load cleaned data

```python
df = pd.read_csv('../data/ai_job_dataset_clean.csv')
```

# Create correlation matrix for numerical features


```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.tight_layout()
plt.savefig('../python_results/correlation_matrix.png')
plt.show()
```

# Create scatter plots: salary vs. years_experience, salary vs. benefits_score


```python
plt.figure(figsize=(6,5))
sns.scatterplot(x='years_experience', y='salary_usd', data=df)
plt.title('Salary vs. Years of Experience')
plt.tight_layout()
plt.savefig('../python_results/salary_vs_years_experience.png')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(x='benefits_score', y='salary_usd', data=df)
plt.title('Salary vs. Benefits Score')
plt.tight_layout()
plt.savefig('../python_results/salary_vs_benefits_score.png')
plt.show()
```

# Perform normality check for salary_usd


```python
shapiro_stat, shapiro_p = stats.shapiro(df['salary_usd'])
print(f"Shapiro-Wilk test statistic: {shapiro_stat:.4f}, p-value: {shapiro_p:.4f}")

plt.figure(figsize=(6,5))
stats.probplot(df['salary_usd'], dist="norm", plot=plt)
plt.title('QQ-plot for Salary')
plt.tight_layout()
plt.savefig('../python_results/qqplot_salary.png')
plt.show()
```

# Create grouped boxplots for categorical features


```python
plt.figure(figsize=(10,5))
sns.boxplot(x='job_title', y='salary_usd', data=df)
plt.title('Salary by Job Title')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../python_results/salary_by_job_title.png')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='company_location', y='salary_usd', data=df)
plt.title('Salary by Company Location')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../python_results/salary_by_company_location.png')
plt.show()
```
