# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
```

# Load cleaned data

```python
df = pd.read_csv('../data/ai_job_dataset_clean.csv')
```

# Create histogram for salary

```python
plt.figure(figsize=(6,5))
sns.histplot(df['salary_usd'], kde=True)
plt.title('Salary Distribution')
plt.tight_layout()
plt.savefig('../python_results/salary_distribution.png')
plt.show()
```

# Create boxplot for salary

```python
plt.figure(figsize=(6,5))
sns.boxplot(x=df['salary_usd'])
plt.title('Salary Boxplot')
plt.tight_layout()
plt.savefig('../python_results/salary_boxplot.png')
plt.show()
```
