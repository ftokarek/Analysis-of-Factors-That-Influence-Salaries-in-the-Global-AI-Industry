# Distribution of the Target Variable (salary_usd)

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, probplot

# Load dataset
df = pd.read_csv("../data/ai_job_dataset.csv")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['salary_usd'], bins=50, kde=True)
plt.title('Salary Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['salary_usd'])
plt.title('Boxplot of Salaries')

plt.tight_layout()
plt.show()
```

# Categorical Boxplots — Salary vs Categorical Features


```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='experience_level', y='salary_usd')
plt.title('Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
df.groupby("job_title")["salary_usd"].median().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.boxplot(x="job_title", y="salary_usd", data=df)
plt.xticks(rotation=45)
plt.title("Salary Distribution by Job Title")
plt.tight_layout()
plt.show()
```

# Correlation Matrix of Numerical Features

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix of Numerical Features")
plt.tight_layout()
plt.show()
```

# Scatterplots — Salary vs Selected Numerical Features

```python
# Example: Remote Ratio
sns.scatterplot(data=df, x='remote_ratio', y='salary_usd')
plt.title('Salary vs Remote Ratio')
plt.xlabel('Remote Ratio (%)')
plt.ylabel('Salary (USD)')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=df, x='years_experience', y='salary_usd')
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (USD)')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=df, x='job_description_length', y='salary_usd')
plt.title('Salary vs Job Description Length')
plt.xlabel('Job Description Length')
plt.ylabel('Salary (USD)')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=df, x='benefits_score', y='salary_usd')
plt.title('Salary vs Benefits Score')
plt.xlabel('Benefits Score')
plt.ylabel('Salary (USD)')
plt.grid(True)
plt.show()
```

# Normality Test for Salary Distribution

```python
from scipy.stats import shapiro, probplot

# Shapiro-Wilk test
stat, p = shapiro(df['salary_usd'])
print(f"Shapiro-Wilk Test p-value: {p:.4f}")

# QQ-Plot
plt.figure(figsize=(6, 6))
probplot(df['salary_usd'], dist="norm", plot=plt)
plt.title("QQ-Plot of Salary Distribution")
plt.grid(True)
plt.show()
```
